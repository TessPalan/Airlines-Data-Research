```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import pathlib
```

```python
flights_1 = pd.read_csv('data/flight_destinations/flights_2019_1.csv')
```

```python
flights_1_cleaned = flights_1.drop(columns='Unnamed: 9')
```

```python
flights_1_cleaned
```

```python
flights_2 = pd.read_csv('data/flight_destinations/flights_2019_2.csv') 
flights_2_cleaned = flights_2.drop(columns='Unnamed: 9')
flights_2_cleaned
```

```python
flights_3 = pd.read_csv('data/flight_destinations/flights_2019_3.csv') 
flights_3_cleaned = flights_3.drop(columns='Unnamed: 9')
flights_3_cleaned
```

```python
flights_4 = pd.read_csv('data/flight_destinations/flights_2019_4.csv') 
flights_4_cleaned = flights_4.drop(columns='Unnamed: 9')
flights_4_cleaned
```

```python
flights_2019 = pd.concat([flights_1, flights_2, flights_3, flights_4], ignore_index=True).drop(columns='Unnamed: 9')
flights_2019
```

```python
flights_2019_renamed = flights_2019.rename(columns={
        "YEAR": "Year", 
        "QUARTER": "Quarter", 
        "ORIGIN": "Origin_Airtport", 
        "ORIGIN_COUNTRY": "Origin_Country", 
        "ORIGIN_STATE_ABR": "Origin_State",
        "DEST": "Destination_Airport", 
        "DEST_COUNTRY": "Destination_Country", 
        "DEST_STATE_ABR": "Destination_State",
        "OPERATING_CARRIER": "Airline_Code"
})
flights_2019_renamed
```

```python
 flights_2019_indexed = flights_2019_renamed.set_index('Year')
```

```python
flights_2019_indexed.to_csv('data/flights_2019.csv')
```
