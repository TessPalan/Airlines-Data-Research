```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import pathlib
```

```python
flights_1 = pd.read_csv('data/flight_destinations/flights_2020_1.csv')
```

```python
flights_1_cleaned = flights_1.drop(columns=['ORIGIN_AIRPORT_ID','PASSENGERS','ORIGIN_WAC','ORIGIN_AIRPORT_SEQ_ID', 'Unnamed: 24','DEST_CITY_MARKET_ID', 'ORIGIN_CITY_MARKET_ID', 'ORIGIN_STATE_FIPS', 'ORIGIN_STATE_NM','DEST_STATE_NM','DEST_AIRPORT_SEQ_ID', 'DEST_AIRPORT_ID','DEST_WAC','TRIP_BREAK','FARE_CLASS', 'DEST_STATE_FIPS'])
```

```python
flights_1_cleaned
```

```python
flights_2 = pd.read_csv('data/flight_destinations/flights_2020_2.csv') 
```

```python
flights_2_cleaned = flights_2.drop(columns=['ORIGIN_AIRPORT_ID','ORIGIN_AIRPORT_SEQ_ID', 'DEST_CITY_MARKET_ID', 'ORIGIN_CITY_MARKET_ID','Unnamed: 13'])
```

```python
flights_2_cleaned
```

```python
flights_3 = pd.read_csv('data/flight_destinations/flights_2020_3.csv') 
```

```python
flights_3_cleaned = flights_3.drop(columns='Unnamed: 9')
```

```python
flights_3_cleaned
```

```python
flights_2020 = pd.concat([flights_1_cleaned, flights_2_cleaned, flights_3_cleaned], ignore_index=True)
flights_2020
```

```python
flights_2020_renamed = flights_2020.rename(columns={
        "YEAR": "Year", 
        "QUARTER": "Quarter", 
        "ORIGIN": "Origin_Airtport", 
        "ORIGIN_COUNTRY": "Origin_Country", 
        "ORIGIN_STATE_ABR": "Origin_State",
        "DEST": "Destination_Airport", 
        "DEST_COUNTRY": "Destination_Country", 
        "DEST_STATE_ABR": "Destination_State",
        "OPERATING_CARRIER": "Airline_Code"
})
flights_2020_renamed
```

```python
 flights_2020_indexed = flights_2020_renamed.set_index('Year')
```

```python
flights_2020_indexed.to_csv('data/flights_2020.csv')
```
